#### Create a bar chart showing the top 12 states with the highest total number of IPC crimes in 2018, using orange bars and dark red labels for the title and axes.

```python

state_crimes = crime_report.groupby("State")["Incidence of ipc ( indian penal code ) crimes"].sum().head(12)


state_crimes.plot(kind='bar', color='orange', figsize=(8,5))
plt.title("Top 12 States by IPC Crimes",color="darkred")
plt.xlabel("State",color="darkred")
plt.ylabel("Number of IPC Crimes",color="darkred")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#### Using a boxplot, analyze the distribution of IPC crime rates across different crime categories in 2018. Which categories show higher variability and which remain consistent?

#### Using a barplot, identify the bottom 10 states with the lowest number of IPC crime incidences in 2018. Which states report the least crimes and how do they compare to each other?"

```python
bottom10 = crime_report.groupby("State")["Incidence of ipc ( indian penal code ) crimes"].sum().sort_values(ascending=True).head(10)
sns.barplot(x=bottom10.index, y=bottom10.values,color="pink")
plt.title("Bottom 10 States by IPC Crime Incidences (2018)")
plt.xlabel("Total Crimes")
plt.xticks(rotation=90)
plt.ylabel("State")
plt.show()
```

#### Which IPC categories contribute most to overall crimes plot using barplot in seaborn?

```python
category_totals = crime_report.groupby("Crime head ipc ( indian penal code ) category")["Incidence of ipc ( indian penal code ) crimes"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x=category_totals.index, y=category_totals.values, color="magenta")
plt.xticks(rotation=90)
plt.title("Total IPC Crimes by Category (2018)")
plt.show()
```

#### Using the dataset, draw a boxplot to visualize the distribution of IPC crimes across different states in 2018.

```python
plt.figure(figsize=(16,8))

sns.boxplot(
    x="State", 
    y="Incidence of ipc ( indian penal code ) crimes",
    data=crime_report,
)

plt.xticks(rotation=90)
plt.title("Distribution of IPC Crimes by State (2018)", fontsize=14, color="darkred")
plt.xlabel("State", fontsize=12,color="darkred")
plt.ylabel("Incidence of IPC Crimes", fontsize=12,color="darkred")
plt.show()
```

#### Using the given dataset, create a violin plot to show the distribution of IPC (Indian Penal Code) crime rates by different categories.

```python
plt.figure(figsize=(12,6))
sns.violinplot(
    x="Crime head ipc ( indian penal code ) category",
    y="Ipc ( indian penal code ) crime rate",
    hue="Crime head ipc ( indian penal code ) category",
    data=crime_report,
    palette="Set2",
    inner="box",
    cut=0,
    legend=False
)

plt.title("Distribution of IPC Crime Rates by Category", fontsize=14, color="darkgreen")
plt.xticks(rotation=45)
plt.ylabel("IPC (Indian Penal Code) Crime Rate", fontsize=12,color="green")
plt.xlabel("Crime Category", fontsize=12,color="green")
plt.show()
```

A line graph is used to display trends or changes over a continuous variable, often time. A pie chart represents data as slices of a circle, showing proportions or percentages of a whole.

Boxplot is helpful to understand the spread of data, showing medians, quartiles, and outliers. Barplot is similar to countplot but usually represents the mean or average of a numerical value across categories. Scatterplot is used to visualize the relationship between two numerical variables by plotting points on a graph. Regplot is an extension of scatterplot that adds a regression line to highlight trends or correlations. Violinplot combines the features of a boxplot with a density curve, providing a deeper view of the data distribution and its variation.
